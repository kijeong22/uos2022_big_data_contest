1. 데이터명 : 연령층별 교통사고
2. 목적 : 2017년부터 2021년까지 각 연령별 pm사고 추이를 보기 위함(ex. '20세이하끼리', '21세이상 30세이하끼리')

```python
# rawData
import pandas as pd
df1 = pd.read_excel('C:\\rawData\\연령층별 교통사고.xls')
df1
```

```python
df1.columns = ['가해운전자 차량종별 대분류','가해운전자 차량종별','기준년도',
'2017_sum','2017_20under','2017_21_30','2017_31_40','2017_41_50','2017_51_60','2017_61_64','2017_65over','2017_unknown',
'2018_sum','2018_20under','2018_21_30','2018_31_40','2018_41_50','2018_51_60','2018_61_64','2018_65over','2018_unknown',
'2019_sum','2019_20under','2019_21_30','2019_31_40','2019_41_50','2019_51_60','2019_61_64','2019_65over','2019_unknown',
'2020_sum','2020_20under','2020_21_30','2020_31_40','2020_41_50','2020_51_60','2020_61_64','2020_65over','2020_unknown',
'2021_sum','2021_20under','2021_21_30','2021_31_40','2021_41_50','2021_51_60','2021_61_64','2021_65over','2021_unknown',
]
df1
```

```python
# PM만 추출 (사용할 것만)
df1_pm = df1.rename(index = {58 : '사고건수', 59 : '사망자수', 60 : '부상자수'}).drop(columns = ['가해운전자 차량종별 대분류','가해운전자 차량종별','기준년도','2017_sum','2018_sum','2019_sum','2020_sum','2021_sum']).iloc[58:61,]
df1_pm
```

```python
df1_pm['2017_61over'] = df1_pm['2017_61_64'] + df1_pm['2017_65over']
df1_pm['2018_61over'] = df1_pm['2018_61_64'] + df1_pm['2018_65over']
df1_pm['2019_61over'] = df1_pm['2019_61_64'] + df1_pm['2019_65over']
df1_pm['2020_61over'] = df1_pm['2020_61_64'] + df1_pm['2020_65over']
df1_pm['2021_61over'] = df1_pm['2021_61_64'] + df1_pm['2021_65over']
df1_pm
```

```python
# 20세이하
df1_pm_20under = df1_pm[['2017_20under','2018_20under','2019_20under','2020_20under','2021_20under']]
df1_pm_20under
```

```python
# 21세이상 30세이하
df1_pm_21_30 = df1_pm[['2017_21_30','2018_21_30','2019_21_30','2020_21_30','2021_21_30']]
df1_pm_21_30
```

```python
# 31세이상 40세이하
df1_pm_31_40 = df1_pm[['2017_31_40','2018_31_40','2019_31_40','2020_31_40','2021_31_40']]
df1_pm_31_40
```

```python
# 41세이상 50세이하
df1_pm_41_50 = df1_pm[['2017_41_50','2018_41_50','2019_41_50','2020_41_50','2021_41_50']]
df1_pm_41_50
```

```python
# 51세이상 60세이하
df1_pm_51_60 = df1_pm[['2017_51_60','2018_51_60','2019_51_60','2020_51_60','2021_51_60']]
df1_pm_51_60
```

```python
# 61세이상
df1_pm_61over = df1_pm[['2017_61over','2018_61over','2019_61over','2020_61over','2021_61over']]
df1_pm_61over
```

- 추가 전처리 : 사고건수만 종합해서 하나의 데이터프레임 만들기

```python
df1_pm_20under.columns = ['2017','2018','2019','2020','2021']
df1_pm_21_30.columns = ['2017','2018','2019','2020','2021']
df1_pm_31_40.columns = ['2017','2018','2019','2020','2021']
df1_pm_41_50.columns = ['2017','2018','2019','2020','2021']
df1_pm_51_60.columns = ['2017','2018','2019','2020','2021']
df1_pm_61over.columns = ['2017','2018','2019','2020','2021']
```

```python
age_new = pd.concat([df1_pm_20under.iloc[0:1,:],df1_pm_21_30.iloc[0:1,:],df1_pm_31_40.iloc[0:1,:],
df1_pm_41_50.iloc[0:1,:],df1_pm_51_60.iloc[0:1,:],df1_pm_61over.iloc[0:1,:]], axis=0)
age_final = age_new.reset_index(drop=True).set_index(pd.Index(['20under','21~30','31~40','41~50','51~60','61over'])).transpose()
age_final
```
