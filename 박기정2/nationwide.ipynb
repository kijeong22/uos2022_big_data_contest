# 전국 시군구별 교통사고 현황

1. 데이터명 : 전국시군구별 교통사고
2. 목적 : 서울뿐만 아니라 전국 각시도에 속한 시군구별 현황을 보여주기 위함
- 사용한다면 광역자치단체별로 분류해서 사용하는 것이 더 나아보임
- 시각화할 분량이 부족할 때 사용

```python
# rawData
import pandas as pd
df4 = pd.read_csv('C:\\rawData\\전국시군구별 교통사고.csv', encoding = 'cp949')
df4
```

```python
# pm data
df4_pm = df4.iloc[:,[0,1,21,50,79,108,137]]
df4_pm.columns = ['자치구명','구분','2017','2018','2019','2020','2021']
df4_pm
```

```python
# 필요한 부분만 추출 -> 최종 pm
df4_pm1 = df4_pm.drop(index=[0,1,2,3,4]) # 합계 제거
a = df4_pm1[['자치구명']].fillna(method='ffill')
b = df4_pm1.drop(columns='자치구명').fillna(0)
df4_pm_fin = pd.concat([a,b],axis=1).reset_index(drop=True)
df4_pm_fin
```
