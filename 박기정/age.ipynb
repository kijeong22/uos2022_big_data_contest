```python
# rawData
import pandas as pd
df1 = pd.read_excel('C:\\rawData\\연령층별 교통사고.xls')
df1
```

```python
df1.columns = ['가해운전자 차량종별 대분류','가해운전자 차량종별','기준년도',
'2017_sum','2017_20under','2017_21_30','2017_31_40','2017_41_50','2017_51_60','2017_61_64','2017_65over','2017_unknown',
'2018_sum','2018_20under','2018_21_30','2018_31_40','2018_41_50','2018_51_60','2018_61_64','2018_65over','2018_unknown',
'2019_sum','2019_20under','2019_21_30','2019_31_40','2019_41_50','2019_51_60','2019_61_64','2019_65over','2019_unknown',
'2020_sum','2020_20under','2020_21_30','2020_31_40','2020_41_50','2020_51_60','2020_61_64','2020_65over','2020_unknown',
'2021_sum','2021_20under','2021_21_30','2021_31_40','2021_41_50','2021_51_60','2021_61_64','2021_65over','2021_unknown',
]
df1
```

```python
# PM만 추출 (사용할 것만)
df1_pm = df1.rename(index = {58 : '사고건수', 59 : '사망자수', 60 : '부상자수'}).drop(columns = ['가해운전자 차량종별 대분류','가해운전자 차량종별','기준년도','2017_sum','2018_sum','2019_sum','2020_sum','2021_sum']).iloc[58:61,]
df1_pm
```

```python
df1_pm['2017_61over'] = df1_pm['2017_61_64'] + df1_pm['2017_65over']
df1_pm['2018_61over'] = df1_pm['2018_61_64'] + df1_pm['2018_65over']
df1_pm['2019_61over'] = df1_pm['2019_61_64'] + df1_pm['2019_65over']
df1_pm['2020_61over'] = df1_pm['2020_61_64'] + df1_pm['2020_65over']
df1_pm['2021_61over'] = df1_pm['2021_61_64'] + df1_pm['2021_65over']
df1_pm
```

```python
# 20세이하
df1_pm_20under = df1_pm[['2017_20under','2018_20under','2019_20under','2020_20under','2021_20under']]
df1_pm_20under
```
