# 서울특별시 전동킥보드 견인 현황

1. 데이터명 : 서울특별시_전동킥보드_견인_현황_20220731
2. 목적 : 이 데이터에서는 다음과 같이 3가지를 도출할예정
- 서울시 내에 어떠한 자치구에서 가장 많은 견인이 발생하는지 -> 전동킥보드가 어떠한 자치구에 가장 많은 도로 혼잡을 야기하는지(단순 바차트로 표현해도 되고, 박재성 팀 자료처럼 서울시 지도에 명암 다르게 해서 표현도 가능할듯)

- 어떠한 견인 유형이 가장 많이 발생하고 있는지 -> 결론에서 단속 및 규제의 방향성을 도출하기 위한 자료

- 실제 견인이 발생하고 있는 자세한 주소 -> 지도에 산점도처럼 표시해서 이렇게나 넓고 랜덤한 지역에서 견인이 발생하고 있음을 보여주기 위함

```python
# rawData
import pandas as pd
df3 = pd.read_csv('C:\\rawData\\서울특별시_전동킥보드_견인_현황_20220731.csv', encoding='cp949')
df3
```

```python
# 어느 구에서 가장 많은 견인이 발생했는지 (2021.07.15 ~ 2022.07.29)
df3_county = df3.groupby(by='구정보').count().reset_index()[['구정보','번호']].rename(columns={'구정보':'자치구명', '번호' : '견인수'}).sort_values('견인수',ascending=False)
df3_county
```

```python
# 어떠한 견인 유형이 가장 많은지
df3_type = df3.groupby(by='유형').count().reset_index()[['유형','번호']].rename(columns={'유형':'견인유형','번호':'견인수'}).sort_values('견인수',ascending=False)
df3_type
```

```python
# 실제 주소 데이터만 추출
df3_address = df3[['주소']]
df3_address
```
