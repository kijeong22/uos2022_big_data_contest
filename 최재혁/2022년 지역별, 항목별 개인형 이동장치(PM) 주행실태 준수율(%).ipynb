```python
import pandas as pd
```

```python
raw_data=pd.read_csv('C:\\rawData\\2022년 지역별, 항목별 개인형 이동장치(PM) 주행실태 준수율(%).csv',encoding='cp949')
```

```python
raw_data
```

```python
data_2022=raw_data[["구분","안전모 착용","주행방향 준수","탑승 인원 준수"]]
data_2021=raw_data[["구분","안전모 착용.1","주행방향 준수.1","탑승 인원 준수.1"]]
```

```python
pm_2021=data_2021[1:10]
pm_2022=data_2022[1:10]
```

```python
pm_2021.replace('-',0,inplace=True)
pm_2021.columns=["구분","안전모 착용","주행방향 준수","탑승 인원 준수"]
pm_2022.replace('-',0,inplace=True)
```

```python
pm_2022
```

```python
pm_merge=pd.concat([pm_2021,pm_2022],axis=1)
pm_merge.columns=["구분","안전모 착용(2021년)","주행방향 준수(2021년)","탑승 인원 준수(2022년)","구분","안전모 착용(2022년)","주행방향 준수(2022년)","탑승 인원 준수(2022년)"]
pm_final=pm_merge[["구분","안전모 착용(2021년)","주행방향 준수(2021년)","탑승 인원 준수(2022년)","안전모 착용(2022년)","주행방향 준수(2022년)","탑승 인원 준수(2022년)"]]
```
