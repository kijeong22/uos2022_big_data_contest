```python
import pandas as pd
```

```python
# local_file_path : 로컬 파일 경로 
local_file_path = '../data/시도별 교통사고.xls' 
colab_file_path =  '/content/gdrive/My Drive/빅데이터경진대회/시도별 교통사고.xls' 
file_path = colab_file_path

df2_raw = pd.read_excel(file_path)
```

```python
df2_raw = df2_raw[['가해운전자 차종별 교통사고', 'Unnamed: 1','Unnamed: 21','Unnamed: 50', 'Unnamed: 79', 'Unnamed: 108', 'Unnamed: 137']].iloc[4:58,:].reset_index(drop = True).replace('-', 0)
```

```python
#시도별 사고건수 사망자수 부상자수 혼합데이터
df2_raw.columns = ["시도", "유형", "2017년", "2018년", "2019년", "2020년", "2021년"]
df_pm_region = df2_raw
df_pm_region
```

```python
#시도별 사고건수 데이터
df_pm_region_acci = df_pm_region[df_pm_region["유형"] == "사고건수"].reset_index(drop = True)
df_pm_region_acci
```

```python
#시도별 사망자수 데이터
df_pm_region_death = df_pm_region[df_pm_region["유형"] == "사망자수"].reset_index(drop = True)
df_pm_region_death
```

```python
#시도별 부상자수 데이터
df_pm_region_injury = df_pm_region[df_pm_region["유형"] == "부상자수"].reset_index(drop = True)
df_pm_region_injury
```
